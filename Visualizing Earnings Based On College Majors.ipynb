#  Visualizing Earnings Based On College   Majors

This project will explore the dataset on the job outcomes of students who graduated from college between 2010 and 2012.(American Community Survey)

```python
%matplotlib inline 
import pandas as pd 
import matplotlib as plt 
```

## Data Exploration 

**Cleaning rows with missing values**

```python
recent_grads = pd.read_csv("recent-grads.csv")  
#exploring the data
#recent_grads.ilc[0]
#recent_grads.head(5)
#recent_grads.tail(5) 
#recent_grads.describe()


# Drop rows with missing values 
raw_data_count = recent_grads.shape[0] 
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0] 
```

**data exploration using scatterplots**

```python
%matplotlib inline 
#data exploration using scatterplots 

#scatter plot of Sample_size and Unemployment_rate 
recent_grads.plot(x= "Sample_size", y="Unemployment_rate", kind = "scatter")


#scatter plot of Full_time and Median  
recent_grads.plot(x= "Full_time", y = "Median", kind = "scatter")

#scatter plot of ShareWomen and Unemployment_rate 
recent_grads.plot(x="ShareWomen", y = "Unemployment_rate", kind = "scatter")

#scatter plot of Men and Median 
recent_grads.plot(x="Men", y ="Median", kind = "scatter")

#scatter Women and Median
recent_grads.plot(x="Women", y="Median", kind ="scatter")
```

## Do students in more popular majors make more money?



```python
%matplotlib inline
#scatterplot of total and median 
recent_grads.plot(x = "Total", y = "Median", kind = "scatter") 
```

since there is no column that shows 'popularity' itself, I used total vs median wage scatter plot. 
Based on the observations, there is weak correleation between the popular majors and median wage. So it is hard to say students in more popular majors(meaning there will be more people in the major) make more money 

## Do students that majored in subjects that were majority female make more money? 

```python
%matplotlib inline 
#scatterplot of Sharewomen and Median
recent_grads.plot(x = "ShareWomen", y = "Median", kind = "scatter")
```

There is a negative correleation between ShareWomen and Median. This means that more female in the major have lower median wage. 
Therefore, it is hard to say that students that majored in subjects that were majority female make more money 

## Is there any link between the number of full-time employees and median salary? 

```python
#scatter plot of Full_time and Median  
recent_grads.plot(x= "Full_time", y = "Median", kind = "scatter")
```

there is weak correlation between full-time employees and median salaray
